```python
import numpy as np
from PIL import Image
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk import word_tokenize
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
words = open('C:/Users/HP/OneDrive/Desktop/DataScience/crooked-house.txt','r').read()
words[1000:3000]
```

```python
tokken_words = word_tokenize(words)
len(tokken_words)
```

```python
fdist = FreqDist(tokken_words)
fdist.most_common(10)
```

```python
words_without_punc = []
for word in tokken_words:
    if word.isalpha():
        words_without_punc.append(word.lower())

len(words_without_punc)
```

```python
stopwords_list = stopwords.words('english')
main_words = []
for word in words_without_punc:
    if word not in stopwords_list:
        main_words.append(word)
        
len(main_words)
```

```python
fdict = FreqDist(main_words)
fdict.most_common(10)
```

```python
clean_words_string = " ".join(main_words)

wordcloud = WordCloud(background_color="lightgreen").generate(clean_words_string)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud)

plt.show()
```
